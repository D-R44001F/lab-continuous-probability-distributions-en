1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers

```python
# your code here
def generate_uniform_numbers(bottom, ceiling, count):
    if bottom >= ceiling:
        raise ValueError("The bottom value must be less than the ceiling value.")
    if count <= 0:
        raise ValueError("The count must be a positive integer.")
    return np.random.uniform(bottom, ceiling, count)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here
def generate_normal_numbers(mean, std_dev, count):
    if count <= 0:
        raise ValueError("The count must be a positive integer.")
    return np.random.normal(loc=mean, scale=std_dev, size=count)

# Generate data
mean_1, std_dev_1 = 10, 1
mean_2, std_dev_2 = 10, 50
count = 1000

data1 = generate_normal_numbers(mean_1, std_dev_1, count)
data2 = generate_normal_numbers(mean_2, std_dev_2, count)

# Plots
plt.figure(figsize=(12, 6))

# First distribution in blue
plt.subplot(1, 2, 1)
sns.histplot(data1, kde=True, color='blue', bins=30)
plt.title(f"Normal Distribution: ={mean_1}, ={std_dev_1}")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Second distribution in gren
plt.subplot(1, 2, 2)
sns.histplot(data2, kde=True, color='green', bins=30)
plt.title(f"Normal Distribution: ={mean_2}, ={std_dev_2}")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def generate_exponential_numbers(mean, size):
    if size <= 0:
        raise ValueError("The size must be a positive integer.")
    
    scale = mean  # Scale parameter for the exponential distribution
    return np.random.exponential(scale, size)

mean = 10
size1 = 10
size2 = 100

sequence1 = generate_exponential_numbers(mean, size1)
sequence2 = generate_exponential_numbers(mean, size2)

# Plot the distributions as histograms
plt.figure(figsize=(14, 6))

# First histogram (size 10)
plt.subplot(1, 2, 1)
sns.histplot(sequence1, bins=100, kde=True, color='blue')
plt.title('Exponential Distribution (Size = 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Second histogram (size 100)
plt.subplot(1, 2, 2)
sns.histplot(sequence2, bins=100, kde=True, color='green')
plt.title('Exponential Distribution (Size = 100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?
The the size makes them appear different, one is much larger, as the mean changes, they both change.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
mean = 10  
lmbda = 1 / mean  
x = 15 

probability = 1 - math.exp(-lmbda * x)

print(f"The probability that a customer will spend less than {x} minutes in the bank is: {probability:.4f}")
```
